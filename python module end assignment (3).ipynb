# 2. Segregate employees based on positions

```python
position_counts = df['Position'].value_counts()
print("\nEmployees segregated based on positions:")
print(position_counts)

# Visualization: Pie chart for position distribution
position_counts.plot(kind='pie', figsize=(8, 6), autopct='%1.1f%%')
plt.title("Distribution of Employees by Position")
plt.axis('equal')
plt.show()
```

Insight:The pie chart visually illustrates the distribution of employees across various positions within the organization.Each segment represents
a different position, with the size of each segment proportional to the number of employees in that role.This visualization provides a quick and intuitive understanding of the relative prevalence of each position, highlighting which roles have a larger or smaller representation in the workforce.
Both teams have a balanced mix of players across different positions, with SG being the most populated and c having the least representation.

# 3. Identify the predominant age group

```python
age_groups = pd.cut(df['Age'], bins=[20, 24, 29, 34, 39, 44], labels=['20-24', '25-29', '30-34', '35-39', '40-44'])
predominant_age_group = age_groups.value_counts().idxmax()
print("\nPredominant age group:", predominant_age_group)

# Visualization: Histogram for age distribution
plt.figure(figsize=(6, 5))
plt.hist(df['Age'], bins=10, edgecolor='black')
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Insight:The predominant age group is 25-29 years old, representing the prime years for professional basketball players.

# 4. Discover which team and position have the highest salary expenditure. 

```python
team_salaries = df.groupby('Team')['Salary'].sum()
team_with_highest_expenditure = team_salaries.idxmax()
print("\nTeam with highest salary expenditure:", team_with_highest_expenditure)
position_salaries = df.groupby('Position')['Salary'].sum()
position_with_highest_expenditure = position_salaries.idxmax()
print("Position with highest salary expenditure:", position_with_highest_expenditure)
# Visualization: Bar chart for team salary expenditure
team_salaries.plot(kind='bar', figsize=(6, 5))
plt.title("Salary Expenditure by Team")
plt.xlabel("Team")
plt.ylabel("Total Salary (in millions)")
plt.show()
```

Insight:The plot illustrates the total salary expenditure for each team, offering a clear comparison of financial allocations across different teams within the organization.

```python
# Visualization: Bar chart for position salary expenditure
position_salaries.plot(kind='bar', figsize=(6, 4))
plt.title("Salary Expenditure by Position")
plt.xlabel("Position")
plt.ylabel("Total Salary (in millions)")
plt.show()
```

Insight:The code generates a bar chart illustrating the total salary expenditure for each position, providing a visual comparison of financial allocations  across different roles within the organization.The center (C) position has the highest total salary expenditure, followed by point guards (PG) and power forwards (PF).
